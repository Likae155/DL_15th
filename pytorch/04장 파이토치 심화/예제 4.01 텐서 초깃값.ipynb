#### 텐서 직접 초기화

수동으로 특정 값을 가진 텐서를 생성하여 초기화

```python
import torch


# FloatTensor를 사용하여 특정 값으로 텐서 초기화
# 3x3 행렬을 직접 지정한 값으로 생성
x = torch.FloatTensor(
    [
        [-0.6577, -0.5797, 0.6360],
        [0.7392, 0.2145, 1.523],
        [0.2432, 0.5662, 0.322]
    ]
)

print("초기화된 텐서:")
print(x)
print(f"\n텐서 shape: {x.shape}")
print(f"텐서 dtype: {x.dtype}")
```
