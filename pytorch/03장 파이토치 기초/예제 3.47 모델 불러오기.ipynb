3. 저장된 모델로 **추론** 수행하기

---

#### 모델 클래스 정의

**주의**: 모델 전체를 저장한 경우, 불러올 때도 동일한 클래스 정의가 필요

```python
import torch
from torch import nn


class CustomModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer = nn.Linear(2, 1)

    def forward(self, x):
        x = self.layer(x)
        return x
```

---

#### 모델 불러오기

**torch.load() 매개변수**
- `map_location`: 모델을 불러올 장치 지정 (GPU↔CPU 간 이동 시 필수)

```python
# 저장된 모델 불러오기
device = "cuda" if torch.cuda.is_available() else "cpu"

# torch.load(): 저장된 모델 전체 불러오기
# map_location: 불러올 장치 지정 (다른 장치에서 저장된 모델도 불러올 수 있음)
model = torch.load("../models/model.pt", map_location=device)
print(model)  # 모델 구조 출력
```

---

#### 불러온 모델로 추론

```python
# 불러온 모델로 예측 수행
with torch.no_grad():
    model.eval()
    inputs = torch.FloatTensor(
        [
            [1 ** 2, 1],    # x=1
            [5 ** 2, 5],    # x=5
            [11 ** 2, 11]   # x=11
        ]
    ).to(device)
    outputs = model(inputs)
    print(outputs)
```
